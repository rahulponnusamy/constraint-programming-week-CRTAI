```python
import pandas as pd
import copy
def create_df(data, list_of_session, list_obj):
  df_list = []
  df_0 = pd.DataFrame(data["activity"])
  for i in range(len(list_of_session)):
    df_copy = copy.deepcopy(df_0)
    df_copy['session_id'] = list_of_session[i] 
    obj_value = list_obj[i] 
    df_copy['obj'] = [obj_value for j in range(134) ]
    solution_id = i
    df_copy['solution_id'] = [solution_id for j in range(134) ]
    df_list.append(df_copy)
  df = pd.concat(df_list)
  return df


df = create_df(data, session_list, objective_list)


    
```
